```python
# Import required Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pathlib import Path
from textblob import TextBlob
```

```python
# Store filepaths to variables
Cancer_disease_prediction = Path("Data/Cancer Disease Prediction.csv")
patient_data = Path("Data/Colorectal Cancer Patient Data.csv")
supplementary_data = Path("Data/supplementary data- deidentified colorectal data.csv")
```

```python
#Read in files
Cancer_disease_prediction_df = pd.read_csv(Cancer_disease_prediction)
patient_data_df = pd.read_csv(patient_data)
supplementary_data_df = pd.read_csv(supplementary_data)
```

## CLEAN SUPPLEMENTARY DATA

```python
# Filter supplementary_data_df 
cleaned_supplementary_df = supplementary_data_df[["Age at diagnosis", "Sex", "ethnicity", "race", "Cancer directed surgery type", 
                                                  "Diagnosed before age 50?", "Histology", "Histologic Grade", "Tumor size (cm)", 
                                                  "Nodes examined", "Nodes positive", "perineural invasion (Y/N)", 
                                                  "Lymphovascular invasion (Y/N)", "TNM Staging",]]
cleaned_supplementary_df.head()
```

```python
# Rename columns cleaned_supplementary_df
cleaned_supplementary_df = cleaned_supplementary_df.rename(columns={"Age at diagnosis":"Age_at_diagnosis", 
                                                                    "Sex":"Gender", "ethnicity": "Ethnicity", "race":"Race",
                                                                    "Cancer directed surgery type": "Surgery_type",
                                                                    "Diagnosed before age 50?": "Diagnosed_before_age_50", 
                                                                    "Histologic Grade":"Histologic_Grade", "Tumor size (cm)":"Tumor_Size_cm",
                                                                    "Nodes examined": "Nodes_examined",
                                                                    "Nodes positive": "Nodes_positive",
                                                                    "perineural invasion (Y/N)": "Perineural_invasion",
                                                                    "Lymphovascular invasion (Y/N)": "Lymphovascular_invasion",                                                                    
                                                                    "TNM Staging":"TNM_Staging",})
cleaned_supplementary_df =cleaned_supplementary_df.dropna()
cleaned_supplementary_df.head()
```

```python
cleaned_supplementary_df["TNM_Staging"].unique()
```

```python
# Map TNM Staging to Dukes Staging

# Convert values in 'TNM Staging' column to uppercase
cleaned_supplementary_df['TNM_Staging'] = cleaned_supplementary_df['TNM_Staging'].str.upper()

# Map TNM Staging to Dukes Staging
tnm_to_dukes_mapping = {
    'I': 'A',
    'IIA': 'B',
    'IIB': 'B',
    'IIC': 'B',
    'IIIA': 'C',
    'IIIB': 'C',
    'IIIC': 'C',
    'IV?': 'D',
    'IVA': 'D',
    'IVB': 'D',
    'IVC': 'D',
}
dukes_stage_list = [tnm_to_dukes_mapping[tnm] for tnm in cleaned_supplementary_df['TNM_Staging']]
cleaned_supplementary_df['Dukes_Stage'] = dukes_stage_list
cleaned_supplementary_df.head()
```

```python
#save dataframe to a CSV file
cleaned_supplementary_df.to_csv('cleaned_supplementary.csv', index=False)
```

```python
# Sample text
a = "Spanish, Nos; Hispan"           # incorrect spelling
print("original text: " + str(a))

# Replace the incorrect term with the correct word
a_corrected = a.replace("Spanish, Nos; Hispan", "Spanish")
print("corrected text: " + str(a_corrected))

```

```python
# Define a function to correct the spelling using TextBlob
def correct_spelling(text):
    return text.replace("Spanish, Nos; Hispan", "Spanish")

# Assuming cleaned_supplementary_df is your DataFrame
# Apply the correction function to the 'Ethnicity' column using apply() and lambda
cleaned_supplementary_df['Ethnicity'] = cleaned_supplementary_df['Ethnicity'].apply(lambda x: correct_spelling(x))

# Print the DataFrame with corrected text
cleaned_supplementary_df.head(20)
```

```python
#save dataframe to a CSV file
cleaned_supplementary_df.to_csv('cleaned_supplementary.csv', index=False)
```

## CLEAN PATIENT DATA

```python
# Filter patient_data_df 
cleaned_patient_df = patient_data_df[["Age (in years)", "Gender", "DFS (in months)", "DFS event", "Adj_Radio", "Adj_Chem", "Dukes Stage"]]
cleaned_patient_df =cleaned_patient_df.dropna()

#convert Age into integers and Months into integers
cleaned_patient_df['Age (in years)'] = cleaned_patient_df['Age (in years)'].astype(int)
cleaned_patient_df['DFS (in months)'] = cleaned_patient_df['DFS (in months)'].astype(int)
cleaned_patient_df.head()
```

```python
# Rename columns cleaned_patient_df
cleaned_patient_df = cleaned_patient_df.rename(columns={"Age (in years)":"Age_at_diagnosis",
                                                        "DFS (in months)":"DFS_in_months",
                                                        "DFS event": "DFS_event",
                                                        "Dukes Stage":"Dukes_Stage"})
cleaned_patient_df.head()
```

```python
#save dataframe to a CSV file
cleaned_patient_df.to_csv('cleaned_patient.csv', index=False)
```
